# **Penjelasan Dataset**

Dataset Analitik Pelanggan Supermarket ini memberikan gambaran mendalam tentang karakteristik dan perilaku pelanggan di sebuah supermarket. Terdapat 29 kolom yang mencakup berbagai aspek kehidupan dan preferensi pelanggan, dibagi 4 kategori :

Seseorang 
- ID: Pengidentifikasi unik pelanggan
- Tahun_Kelahiran: Tahun lahir Pelanggan
- Pendidikan: Tingkat pendidikan pelanggan
- Marital_Status: Status perkawinan pelanggan
-  Pendapatan: Pendapatan rumah tangga tahunan pelanggan
-  Kidhome: Jumlah anak di rumah pelanggan
- Teenhome: Jumlah remaja di rumah pelanggan
-  Dt_Customer: Tanggal pendaftaran pelanggan pada perusahaan
- Keterkinian: Jumlah hari sejak pembelian terakhir pelanggan
- Komplain: 1 jika pelanggan mengeluh dalam 2 tahun terakhir, 0 sebaliknya

Produk
-  MntWines: Jumlah yang dibelanjakan untuk wine dalam 2 tahun terakhir
-  MntFruits: Jumlah yang dibelanjakan untuk buah-buahan dalam 2 tahun terakhir
-  MntMeatProducts: Jumlah yang dibelanjakan untuk daging dalam 2 tahun terakhir
- MntFishProducts: Jumlah yang dibelanjakan untuk ikan dalam 2 tahun terakhir
- MntSweetProducts: Jumlah yang dibelanjakan untuk permen dalam 2 tahun terakhir
- MntGoldProds: Jumlah yang dibelanjakan untuk emas dalam 2 tahun terakhir

Promosi
- NumDealsPurchases: Jumlah pembelian yang dilakukan dengan diskon
- AcceptedCmp1: 1 jika pelanggan menerima tawaran di kampanye pertama, 0 jika tidak
- AcceptedCmp2: 1 jika pelanggan menerima tawaran di kampanye ke-2, 0 sebaliknya
- AcceptedCmp3: 1 jika pelanggan menerima tawaran di kampanye ke-3, 0 sebaliknya
- AcceptedCmp4: 1 jika pelanggan menerima tawaran di kampanye ke-4, 0 sebaliknya
- AcceptedCmp5: 1 jika pelanggan menerima tawaran pada kampanye ke-5, 0 sebaliknya
- Respons: 1 jika pelanggan menerima tawaran pada kampanye terakhir, 0 sebaliknya

Tempat
- NumWebPurchases: Jumlah pembelian yang dilakukan melalui situs web perusahaan
- NumCatalogPurchases: Jumlah pembelian yang dilakukan menggunakan katalog
- NumStorePurchases: Jumlah pembelian yang dilakukan langsung di toko
- NumWebVisitsMonth: Jumlah kunjungan ke situs web perusahaan dalam sebulan terakhir

**Tidak Terpengaruh Nilai Ekstrim** , akan lebih aman karena median hanya ngambil nilai tengah, sehingga tidak terpengaruh oleh outlier atau nilai ekstrim. Ini membuat median lebih stabil dan baik untuk distribusi data yang mungkin memiliki outlier.

### D. Penanganan Outlier

```python
Q1 = df['Income'].describe()['25%']
Q3 = df['Income'].describe()['75%']

IQR = Q3 - Q1

batas_bawah = Q1 - (1.5 * IQR)
batas_atas = Q3 + (1.5 * IQR)

df[(df['Income'] < batas_bawah) | (df['Income'] > batas_atas)]
```

```python
sns.boxplot(df['Income']);
```

```python
# Menghapus outlier
data = df[~((df['Income'] < batas_bawah) | (df['Income'] > batas_atas))]
data
```

```python
# Hasil Outlier
sns.boxplot(data['Income']);
```

Ini adalah outlier setelah dihapus

# Analysis Data

### **1. Siapa pelanggan utama supermarket?**

```python
# Menambahkan kolom usia pelanggan berdasarkan tahun lahir
data['Age'] = 2024 - data['Year_Birth']

# Visualisasi distribusi usia pelanggan
plt.figure(figsize=(10, 6))
sns.histplot(data['Age'], kde=True, bins=30, color = 'Green')
plt.title('Distribusi Usia Pelanggan')
plt.xlabel('Usia')
plt.ylabel('Frekuensi')
plt.show()

# Visualisasi distribusi status perkawinan
plt.figure(figsize=(10, 6))
sns.countplot(y='Marital_Status', data=data,palette='plasma', order=data['Marital_Status'].value_counts().index)
plt.title('Distribusi Status Perkawinan Pelanggan')
plt.xlabel('Frekuensi')
plt.ylabel('Status Perkawinan')
plt.show()

# Visualisasi distribusi tingkat pendidikan
plt.figure(figsize=(10, 6))
sns.countplot(y='Education', data=data, order=data['Education'].value_counts().index)
plt.title('Distribusi Tingkat Pendidikan Pelanggan')
plt.xlabel('Frekuensi')
plt.ylabel('Tingkat Pendidikan')
plt.show()
```

**Insight 1: Profil Pelanggan Utama**
- Usia: Mayoritas pelanggan berusia 30-50 tahun.
- Status Perkawinan: Banyak yang sudah menikah.
- Pendidikan: Sebagian besar berpendidikan tinggi (sarjana atau lebih).

### **2. Produk apa yang paling banyak dibeli oleh pelanggan?**

```python
# Menghitung total pembelian untuk setiap kategori produk
product_totals = data[['MntWines', 'MntFruits', 'MntMeatProducts', 'MntFishProducts', 'MntSweetProducts', 'MntGoldProds']].sum()

# Membuat daftar kategori produk
categories = ['Wine', 'Fruits', 'Meat Products', 'Fish Products', 'Sweet Products', 'Gold Products']

# Visualisasi total pembelian untuk setiap kategori produk
plt.figure(figsize=(12, 8))
sns.barplot(x=product_totals.values, y=categories, palette='viridis')

# Menambahkan judul dan label
plt.title('Total Pembelian untuk Setiap Kategori Produk', fontsize=16)
plt.xlabel('Total Pembelian', fontsize=14)
plt.ylabel('Kategori Produk', fontsize=14)

# Menambahkan anotasi pada setiap bar
for index, value in enumerate(product_totals):
    plt.text(value, index, f'{value:.0f}', va='center', ha='left', fontsize=12)

plt.show()
```

**Insight yang Didapat**
- Wine adalah produk yang paling populer dan paling banyak dibeli oleh pelanggan.
- Produk daging (Meat Products) juga sangat populer dengan jumlah pembelian yang signifikan.
- Produk emas (Gold Products) memiliki pembelian yang relatif tinggi dibandingkan dengan produk ikan, manis, dan buah-buahan.

**Rekomendasi**
- Fokus pada Produk Populer: Meningkatkan stok dan promosi untuk produk wine dan meat products karena permintaannya tinggi.
- Optimalisasi Produk Lainnya: Lakukan promosi yang lebih agresif untuk produk dengan pembelian lebih rendah seperti sweet products dan fruits untuk meningkatkan penjualan di kategori tersebut.
- Analisis Preferensi Pelanggan:  Lakukan analisis mendalam pada pelanggan yang membeli produk terbanyak untuk mengetahui siapa mereka dan bagaimana pola belanja mereka.

### **3. Bagaimana tingkat respons pelanggan terhadap kampanye pemasaran?**

```python
# Menghitung persentase respons pelanggan untuk setiap kampanye pemasaran
campaign_columns = ['AcceptedCmp1', 'AcceptedCmp2', 'AcceptedCmp3', 'AcceptedCmp4', 'AcceptedCmp5', 'Response']
campaign_responses = data[campaign_columns].mean() * 100

# Membuat daftar nama kampanye
campaign_names = ['Kampanye 1', 'Kampanye 2', 'Kampanye 3', 'Kampanye 4', 'Kampanye 5', 'Kampanye Terakhir']

# Visualisasi persentase respons pelanggan untuk setiap kampanye pemasaran
plt.figure(figsize=(12, 8))
sns.barplot(x=campaign_names, y=campaign_responses.values, palette='viridis')

# Menambahkan judul dan label
plt.title('Persentase Respons Pelanggan untuk Setiap Kampanye Pemasaran', fontsize=16)
plt.xlabel('Kampanye Pemasaran', fontsize=14)
plt.ylabel('Persentase Respons (%)', fontsize=14)

# Menambahkan anotasi pada setiap bar
for index, value in enumerate(campaign_responses):
    plt.text(index, value + 0.5, f'{value:.2f}%', ha='center', fontsize=12)

# Memutar label x agar lebih mudah dibaca
plt.xticks(rotation=45)

plt.show()
```

**Insight 3: Respons terhadap Kampanye Pemasaran**
- Kampanye terakhir paling efektif dengan hampir 15% respons.
- Kampanye 2 paling tidak efektif dengan hanya 1.34% respons.
- Kampanye 3, 4, dan 5 memiliki efektivitas yang cukup seimbang.

**Rekomendasi**

Persentase Respons Pelanggan untuk Setiap Kampanye Pemasaran:
- Tingkatkan Analisis Kampanye: Lakukan analisis lebih mendalam terhadap kampanye terakhir untuk memahami faktor kesuksesannya dan tiru elemen-elemen tersebut dalam kampanye mendatang.
- Evaluasi Kampanye 2: Pertimbangkan untuk merevisi atau menghentikan Kampanye 2 karena efektivitasnya yang sangat rendah. Cari tahu alasan di balik rendahnya respons.

### **4. Bagaimana hubungan antara pendapatan dan pola pembelian?**

```python
# Korelasi antara pendapatan dan berbagai kategori pembelian
correlation_matrix = data[['Income', 'MntWines', 'MntFruits', 'MntMeatProducts', 'MntFishProducts', 'MntSweetProducts', 'MntGoldProds']].corr()

# Visualisasi korelasi antara pendapatan dan kategori pembelian
plt.figure(figsize=(10, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Korelasi antara Pendapatan dan Kategori Pembelian')
plt.show()
```

**Insight:**
- Pendapatan (Income) memiliki korelasi yang kuat dengan jumlah pembelian wine (MntWines) dengan koefisien sebesar 0.73. Ini menunjukkan bahwa semakin tinggi pendapatan seseorang, semakin besar kemungkinan mereka untuk membeli wine dalam jumlah yang lebih banyak.

- Pembelian produk daging (MntMeatProducts) juga memiliki korelasi yang cukup kuat dengan pendapatan, dengan koefisien sebesar 0.69. Artinya, pendapatan yang lebih tinggi juga berkorelasi dengan pembelian produk daging yang lebih banyak.

- Korelasi dengan pembelian produk lainnya seperti buah-buahan (MntFruits), produk ikan (MntFishProducts), permen (MntSweetProducts), dan emas (MntGoldProds) juga positif tetapi lebih rendah dibandingkan wine dan daging.

- Korelasi yang terendah ada pada pembelian produk emas (MntGoldProds), dengan nilai koefisien sebesar 0.42. Ini menunjukkan bahwa pendapatan memiliki pengaruh yang lebih kecil terhadap jumlah pembelian produk emas dibandingkan kategori produk lainnya
